# Detect Chart Type

```python
from tensorflow.keras.saving import load_model
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def detectChartType(model_path, image_path, width=224, height=224):

    # Load and preprocess the image
    image = load_img(image_path, target_size=(width, height))

    # Add a batch dimension to the image
    image = np.expand_dims(image, axis=0)
    image = image.astype('float32') / 255.0  # Normalize

    # Load the model
    model = load_model(model_path)

    # Make predictions
    predictions = model(image, training=False)
    max_index = np.argmax(predictions)  # Get the index of the max score
    class_names = ["vbar_categorical", "hbar_categorical", "dot_line", "line"]

    return class_names[max_index]
```

# Detect Chart Elements for Scatter Plots



```python
import torch
import detectron2
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
import cv2
import matplotlib.pyplot as plt

DATASET_NAME = "scatter_plot_elements"

MetadataCatalog.get(DATASET_NAME).thing_classes = ["title", "canvas", "x-labels",
                                                   "y-labels", "x-title", "y-title",
                                                   "legend", "scale", "other text",
                                                   "quantity"]

# Load Detectron2 configuration
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.WEIGHTS = SCATTER_PLOT_ELEMENT_DETECTION_MODEL_PATH
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 10
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7
cfg.MODEL.DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
import random

def detectChartElements(IMAGE_PATH):
    if detectChartType(CHART_TYPE_DETECTION_MODEL_PATH, IMAGE_PATH) != "dot_line":
        return "Not a Scatter Plot"

    # Initialize predictor
    predictor = DefaultPredictor(cfg)

    # Load the image
    image = cv2.imread(IMAGE_PATH)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Run inference
    outputs = predictor(image)

    # Extract bounding boxes, labels, and scores
    instances = outputs["instances"].to("cpu")
    boxes = instances.pred_boxes.tensor.numpy()
    labels = instances.pred_classes.numpy()
    scores = instances.scores.numpy()

    # Metadata for class names
    metadata = MetadataCatalog.get(DATASET_NAME)
    class_names = metadata.get("thing_classes", [])

    # Dictionary to store the best element per label
    detected_elements = {}
    colors = {}

    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        cropped_img = (x1, y1, x2, y2)
        label = class_names[labels[i]] if labels[i] < len(class_names) else f"class_{labels[i]}"

        # Store only the highest confidence bounding box per label
        if label not in detected_elements or scores[i] > detected_elements[label][0][1]:
            detected_elements[label] = [(cropped_img, scores[i])]
            colors[label] = [random.random(), random.random(), random.random()]

    # Visualize the final results with only the best bounding boxes
    v = Visualizer(image, metadata, scale=1.2)
    for label, box in detected_elements.items():
        x1, y1, x2, y2 = box[0][0]
        color = colors[label]
        v.draw_box([x1, y1, x2, y2], edge_color=color)
        v.draw_text(f"{label} ({box[0][1]:.2f})", [x1, y1], font_size=10, color=color)

    # Remove confidence scores from final output
    detected_elements = {label: [box[0][0]] for label, box in detected_elements.items()}

    # Display the final image
    plt.figure(figsize=(10, 10))
    plt.imshow(v.output.get_image())
    plt.axis("off")
    plt.show()

    return detected_elements
```

```python
detectedElements = detectChartElements(IMAGE_PATH)
print(detectedElements)
```

```python
def rotate_image(image, angle):
    """Rotate image by a given angle and return the rotated image."""
    h, w = image.shape[:2]
    center = (w // 2, h // 2)

    # Get rotation matrix
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Compute new bounding box size
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])
    new_w = int(h * sin + w * cos)
    new_h = int(h * cos + w * sin)

    # Adjust transformation matrix to move image to the center
    M[0, 2] += (new_w - w) / 2
    M[1, 2] += (new_h - h) / 2

    # Rotate the image with new bounds
    rotated = cv2.warpAffine(image, M, (new_w, new_h))

    return rotated, M
```

```python
def rotate_roi(roi_x1, roi_y1, roi_x2, roi_y2, M):
  # Define the four corner points of the ROI in the original image
  roi_corners = np.array([
      [roi_x1, roi_y1],  # Top-left
      [roi_x2, roi_y1],  # Top-right
      [roi_x2, roi_y2],  # Bottom-right
      [roi_x1, roi_y2]   # Bottom-left
  ], dtype=np.float32)

  # Convert to homogeneous coordinates
  ones = np.ones((4, 1))
  roi_corners_homogeneous = np.hstack([roi_corners, ones])

  # Transform ROI coordinates using the rotation matrix M
  transformed_corners = np.dot(M, roi_corners_homogeneous.T).T

  # Get bounding box of transformed ROI
  x_min = int(min(transformed_corners[:, 0]))
  y_min = int(min(transformed_corners[:, 1]))
  x_max = int(max(transformed_corners[:, 0]))
  y_max = int(max(transformed_corners[:, 1]))

  # Ensure the new ROI is within image bounds
  x_min, y_min = max(0, x_min), max(0, y_min)
  x_max, y_max = min(image.shape[1], x_max), min(image.shape[0], y_max)

  return x_min, y_min, x_max, y_max
```

```python
def inverse_transform_box(box, M):
    """Transform bounding box coordinates back to original image."""
    box = np.array(box, dtype=np.float32)
    ones = np.ones((4, 1))
    box = np.hstack([box, ones])  # Convert to homogeneous coordinates
    inv_M = cv2.invertAffineTransform(M)
    transformed_box = np.dot(inv_M, box.T).T
    return transformed_box[:, :2].astype(int)
```

```python
def detect_text_in_rois(image_path, rois, width_threshold_pct=0.02, height_threshold_pct=0.1):
    # Load the image
    original = cv2.imread(image_path)
    if original is None:
        print("Error: Unable to load image.")
        return []

    best = []
    best_confidence = 0

    for angle in [0, 90, 180, 270]:
        detected_text = []  # Store (bounding_box, text)
        totalConfidence = 0
        count = 0

        for roi_x1, roi_y1, roi_x2, roi_y2 in rois:
            image = original[roi_y1:roi_y2, roi_x1:roi_x2]
            roi, M = rotate_image(image, angle)

            # Run OCR on the cropped ROI
            results = ocr.ocr(roi, cls=True)

            if results is None:
                continue

            for res in results:
                if res is None:
                    continue

                for line in res:
                    if line is None:
                        continue

                    box, (text, confidence) = line  # Get bounding box and text

                    if confidence < 0.5:
                      continue

                    # Transform bounding box back to original image coordinates
                    original_box = inverse_transform_box(box, M)

                    # Convert bounding box to global coordinates
                    adjusted_box = [
                        (int(pt[0]) + roi_x1, int(pt[1]) + roi_y1) for pt in original_box
                    ]

                    totalConfidence += confidence
                    count += 1
                    detected_text.append((adjusted_box, text))

        if count > 0:
          totalConfidence /= count

        if totalConfidence > best_confidence:
            best_confidence = totalConfidence
            best = detected_text

    for (adjusted_box, text) in best:
      # Draw bounding box and text for visualization
      cv2.polylines(original, [np.array(adjusted_box, np.int32)], isClosed=True, color=(0, 255, 0), thickness=2)
      cv2.putText(original, text, (adjusted_box[0][0], adjusted_box[0][1] - 10),
                  cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show the image with detected text
    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    return best  # Return list of (bounding_box, text)
```

```python
def is_overlapping(box1, box2):
    """Check if two bounding boxes overlap and draw them using plt."""
    x1_min, y1_min, x1_max, y1_max = box1
    x2_max, y2_max, x2_min, y2_min = box2

    # Ensure min and max are correctly ordered
    x1_min, x1_max = min(x1_min, x1_max), max(x1_min, x1_max)
    y1_min, y1_max = min(y1_min, y1_max), max(y1_min, y1_max)
    x2_min, x2_max = min(x2_min, x2_max), max(x2_min, x2_max)
    y2_min, y2_max = min(y2_min, y2_max), max(y2_min, y2_max)

    return not (x1_max <= x2_min or x1_min >= x2_max or y1_max <= y2_min or y1_min >= y2_max)
```

# Find Corresponding Labels in the Canvas

```python
def get_hsv_range(bgr, tol=10):
    """ Convert BGR to HSV range for masking. """
    color = np.uint8([[bgr]])  # Convert to NumPy array
    hsv_color = cv2.cvtColor(color, cv2.COLOR_BGR2HSV)[0][0]

    lower = np.array([max(hsv_color[0] - tol, 0), 50, 50])  # Lower HSV bound
    upper = np.array([min(hsv_color[0] + tol, 179), 255, 255])  # Upper HSV bound
    return lower, upper
```

```python
def detect_colored_circles(image, roi, mapped_legends):
    """ Detect colored circles in a given Region of Interest (ROI). """
    x1, y1, x2, y2 = roi  # ROI (top-left, bottom-right)
    roi_img = image[y1:y2, x1:x2]  # Extract the region of interest

    hsv_roi = cv2.cvtColor(roi_img, cv2.COLOR_BGR2HSV)
    detected_positions = {}

    for label, bgr in mapped_legends.items():
        lower, upper = get_hsv_range(bgr)
        mask = cv2.inRange(hsv_roi, lower, upper)  # Threshold image

        # Find contours
        cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        detected_circles = []
        for c in cnts:
            perimeter = cv2.arcLength(c, True)
            approx = cv2.approxPolyDP(c, 0.04 * perimeter, True)

            # If contour has enough points, assume it's a circle
            if len(approx) >= 5:
                (cx, cy), radius = cv2.minEnclosingCircle(c)
                absolute_x = x1 + int(cx)
                absolute_y = y1 + int(cy)
                detected_circles.append((absolute_x, absolute_y, int(radius)))

        # Store detected circles for the current label
        detected_positions[label] = detected_circles

    return detected_positions
```

```python
# Draw detected circles on the image
def draw_detected_circles(image, detected_positions):
    count = 1  # Counter for numbering detected circles

    for label, circles in detected_positions.items():
        print(label, count)
        for (cx, cy, radius) in circles:
            cv2.circle(image, (cx, cy), radius, (0, 255, 0), 2)  # Draw green circle
            cv2.putText(image, str(count), (cx, cy - radius - 5), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (0, 0, 255), 1, cv2.LINE_AA)  # Label with number
        count += 1  # Increment counter
```

```python
def extract_center(points):
    """ Extracts the center x or y coordinate from a bounding box. """
    x_coords, y_coords = zip(*points)
    return round(sum(x_coords) / len(x_coords)), round(sum(y_coords) / len(y_coords))
```

```python
import numpy as np
import re

def smart_convert(value):
    """
    Attempts to convert a string to a float.
    Handles common numeric formats and strips non-numeric characters if necessary.
    Returns np.nan if conversion fails.
    """
    try:
        return float(value)
    except ValueError:
        # Try to extract float using regex
        match = re.search(r"[-+]?\d*\.\d+|\d+", str(value))
        if match:
            return float(match.group())
        return np.nan
```

```python
import numpy as np
from scipy.interpolate import interp1d

def map_coordinates(detected_points, x_labels, y_labels):
    # Extract center positions from bounding boxes with robust float conversion
    x_positions = [(extract_center(box)[0], smart_convert(value)) for box, value in x_labels]
    y_positions = [(extract_center(box)[1], smart_convert(value)) for box, value in y_labels]

    # Filter out any positions where conversion failed (np.nan)
    x_positions = [(x, v) for x, v in x_positions if not np.isnan(v)]
    y_positions = [(y, v) for y, v in y_positions if not np.isnan(v)]

    if len(x_positions) < 2 or len(y_positions) < 2:
        raise ValueError("At least two valid coordinate labels are required for piecewise interpolation.")

    # Sort by pixel position to ensure monotonicity
    x_positions.sort()
    y_positions.sort()

    x_pixels, x_values = zip(*x_positions)
    y_pixels, y_values = zip(*y_positions)

    # Create piecewise linear interpolators
    x_to_value = interp1d(x_pixels, x_values, kind='linear', fill_value='extrapolate')
    y_to_value = interp1d(y_pixels, y_values, kind='linear', fill_value='extrapolate')

    # Convert detected coordinates to (x-mapped value, y-mapped value)
    mapped_points = {
        category: [(round(float(x_to_value(x)), 2), round(float(y_to_value(y)), 2)) for x, y, _ in points]
        for category, points in detected_points.items()
    }

    return mapped_points
```
